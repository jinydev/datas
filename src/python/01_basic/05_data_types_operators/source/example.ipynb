```python
# 3.1.5 complex: 복소수 생성
c1 = 3 + 4j
c2 = complex(1, -2) # 내장 함수를 사용한 생성 (1 - 2j)

print(c1, "는", type(c1), "형식입니다.")

# 복소수의 사칙연산 (파이썬이 알아서 처리함)
print("덧셈:", c1 + c2)
print("곱셈:", c1 * c2)

# 복소수 객체의 유용한 내장 속성과 메서드
print("실수부만 추출:", c1.real)
print("허수부만 추출:", c1.imag)
print("켤레복소수(허수부 부호 반전):", c1.conjugate())
```

```python
import sys
import matplotlib.pyplot as plt

# 다양한 타입의 데이터 메모리 크기 측정 (Bytes)
types_data = {
    "int(0)": sys.getsizeof(0),
    "int(1억)": sys.getsizeof(100000000),
    "float(3.14)": sys.getsizeof(3.14),
    "bool(True)": sys.getsizeof(True),
    "str('A')": sys.getsizeof('A'),
    "str('Hello World')": sys.getsizeof('Hello World')
}

labels = list(types_data.keys())
sizes = list(types_data.values())

plt.figure(figsize=(10, 5))
plt.barh(labels, sizes, color='teal')
plt.title("Memory Size of Python Data Types (Bytes)")
plt.xlabel("Size in Bytes")
for index, value in enumerate(sizes):
    plt.text(value, index, str(value))
plt.show()
```
