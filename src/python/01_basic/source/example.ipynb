```python
a = 10
b = 3

print(a + b)  # 덧셈: 13
print(a - b)  # 뺄셈: 7
print(a * b)  # 곱셈: 30
print(a / b)  # 나눗셈: 3.3333... (일반적인 나눗셈, 결과는 항상 실수(float)로 반환됩니다)
print(a // b) # 몫: 3 (소수점을 버리고 정수 몫만 반환하는 정수 나눗셈입니다)
print(a % b)  # 나머지: 1 (매우 중요! 특정한 배수나 짝수/홀수를 판별할 때 핵심적으로 쓰입니다)
print(a ** b) # 거듭제곱: 1000 (10의 3승을 의미합니다)
```

```python
text = "ABCDEFG"

# 3.1 인덱싱 (Indexing): 특정 '한 글자' 추출
print(text[0])   # 첫 번째 글자 (인덱스 0): 'A'
print(text[1])   # 두 번째 글자 (인덱스 1): 'B'
print(text[-1])  # 마지막 글자 (인덱스 순서를 거꾸로 세면 -1부터 시작): 'G'

# 3.1 슬라이싱 (Slicing): 특정 '구간' 추출 [시작위치 : 끝위치값(포함 안 됨)]
print(text[0:3]) # 0번 인덱스부터 3번 '직전'인 2번(0, 1, 2)까지만 잘라냅니다: 'ABC'
print(text[3:])  # 3번 인덱스부터 문자열의 끝까지 모두 가져옵니다: 'DEFG'
```
