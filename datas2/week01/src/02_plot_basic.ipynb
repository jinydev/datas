# 02. 데이터의 기하학
좌표계를 이해하고 기본적인 도형을 그려봅니다.

```python
import matplotlib.pyplot as plt

# 9. 사각형 그리기
# 순서: 좌하단 -> 좌상단 -> 우상단 -> 우하단 -> 좌하단
x_square = [1, 1, 2, 2, 1]
y_square = [1, 2, 2, 1, 1]

# 15. Python 문법 집중: 함수의 인자
plt.figure(figsize=(6, 6))

# 사각형 그리기
plt.plot(x_square, y_square, marker='s', color='blue', label='Square')

# 19. 응용 예제: 집 그리기
# 지붕 (삼각형)
x_roof = [1, 1.5, 2]
y_roof = [2, 2.8, 2]
plt.plot(x_roof, y_roof, marker='^', color='red', label='Roof')

# 꾸미기
plt.title("A Simple House Construction")
plt.grid(True)
plt.legend()
plt.xlim(0.5, 2.5)
plt.ylim(0.5, 3.5)
plt.show()
```
