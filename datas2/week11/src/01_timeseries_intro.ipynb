# 01. 시계열 데이터 기초 (Time Series)
이동 평균, 시계열 분해(Trend/Seasonal/Resid), ACF/PACF 실습.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# 가상의 시계열 데이터 생성
dates = pd.date_range('2023-01-01', periods=100)
data = pd.DataFrame(data=np.random.randn(100).cumsum(), index=dates, columns=['Price'])

# 이동 평균
data['MA'] = data['Price'].rolling(window=10).mean()
data.plot(figsize=(12, 5))
plt.title('Stock Price with Moving Average')
plt.show()
```

```python
# 시계열 분해
decomp = seasonal_decompose(data['Price'], model='additive', period=12)
decomp.plot()
plt.show()
```

```python
# 차분 (Differencing) - 정상성 만들기
data['Diff'] = data['Price'].diff()
data['Diff'].plot()
plt.title('Differenced Data (Stationary)')
plt.show()
```
