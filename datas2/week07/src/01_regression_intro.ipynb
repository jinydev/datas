# 01. 선형 회귀 (Linear Regression)
사이킷런을 이용한 단순 선형 회귀 및 다항 회귀 실습.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 데이터 생성
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# 모델 학습
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# 결과 시각화
X_new = np.array([[0], [2]])
y_predict = lin_reg.predict(X_new)

plt.scatter(X, y)
plt.plot(X_new, y_predict, "r-", linewidth=2)
plt.title(f"y = {lin_reg.coef_[0][0]:.2f}x + {lin_reg.intercept_[0]:.2f}")
plt.show()
```

```python
# 결정 계수 (R^2) 확인
lin_reg.score(X, y)
```
