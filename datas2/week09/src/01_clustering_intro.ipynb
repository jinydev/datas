# 01. 비지도 학습 (Clustering & PCA)
K-Means, 엘보우 기법, PCA 시각화 실습.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# 데이터 로드 및 스케일링
iris = load_iris()
X = iris.data
X_scaled = StandardScaler().fit_transform(X)
```

```python
# 엘보우 기법
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 10), inertia, 'o-')
plt.title('Elbow Method')
plt.show()
```

```python
# PCA 시각화
kmeans = KMeans(n_clusters=3)
pred = kmeans.fit_predict(X_scaled)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=pred)
plt.title(f'Silhouette Score: {silhouette_score(X_scaled, pred):.2f}')
plt.show()
```
