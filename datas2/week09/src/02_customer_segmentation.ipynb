# 02. 고객 세분화 프로젝트 (Customer Segmentation)
쇼핑몰 고객 데이터를 이용한 RFM 분석 유사 실습.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 가상의 쇼핑몰 데이터 생성 (소득, 지출점수)
import numpy as np
X1 = np.random.normal(25, 5, (50, 2))
X2 = np.random.normal(80, 5, (50, 2))
X3 = np.random.normal(50, 10, (50, 2))
X = np.vstack([X1, X2, X3])
df = pd.DataFrame(X, columns=['Income', 'Score'])

# 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
# K-Means 학습
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# 결과 시각화
sns.scatterplot(data=df, x='Income', y='Score', hue='Cluster', palette='viridis')
plt.title('Customer Segments')
plt.show()
```

```python
# 마케팅 인사이트 도출
df.groupby('Cluster').mean()
```
