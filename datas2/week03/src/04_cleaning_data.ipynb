# 04. 데이터 정제 (Cleaning)
결측치(NaN) 처리와 중복 제거 실습.

```python
import pandas as pd
import numpy as np

# 더러운 데이터 생성
data = {
    'group': ['A', 'A', 'B', np.nan, 'B'],
    'value': [10, np.nan, 30, 40, 30] # 마지막은 중복
}
df = pd.DataFrame(data)
df
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 평균으로 채우기
mean_val = df['value'].mean()
df['value'] = df['value'].fillna(mean_val)
df
```

```python
# 중복 제거
df.drop_duplicates()
```
