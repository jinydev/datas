# 03. 데이터 가공
파생 변수 생성, 날짜 처리, apply 함수 활용하기.

```python
import pandas as pd

# 샘플 데이터
data = {
    '주문일자': ['2023-01-01', '2023-01-02', '2023-01-05', '2023-02-10'],
    '금액': [50000, 150000, 30000, 200000]
}
df = pd.DataFrame(data)

# 13. 날짜형 변환
df['주문일자'] = pd.to_datetime(df['주문일자'])

# 14. 요일 추출
df['요일'] = df['주문일자'].dt.day_name()
df
```

```python
# 6. Lambda와 apply를 이용한 파생 변수
# 금액이 10만원 이상이면 True, 아니면 False
df['고액주문'] = df['금액'].apply(lambda x: x >= 100000)
df
```

```python
# 8. 컬럼 삭제
df.drop('금액', axis=1) # 원본 유지, 결과만 출력
```
