# 05. 데이터 집계 (GroupBy)
그룹별 통계량을 계산하고 피벗 테이블 만들기.

```python
import pandas as pd

# 샘플 데이터
data = {
    '지점': ['강남점', '강남점', '강남점', '홍대점', '홍대점', '홍대점'],
    '요일': ['월', '화', '수', '월', '화', '수'],
    '상품': ['물', '커피', '물', '커피', '물', '커피'],
    '매출': [10000, 40000, 15000, 100000, 30000, 90000]
}
df = pd.DataFrame(data)
df
```

```python
# 지점별 총 매출
df.groupby('지점')[['매출']].sum()
```

```python
# 요일별 상품별 매출 평균
df.groupby(['요일', '상품'])[['매출']].mean()
```

```python
# 피벗 테이블
pd.pivot_table(df, index='지점', columns='요일', values='매출', aggfunc='sum')
```
