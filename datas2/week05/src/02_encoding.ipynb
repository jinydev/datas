# 02. 범주형 데이터 변환 (Encoding)
Label Encoding과 One-Hot Encoding 실습.

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

df = sns.load_dataset('titanic')
data = df[['sex', 'embarked', 'class']].dropna()

# Pandas Get Dummies (가장 간편)
pd.get_dummies(data, columns=['sex', 'embarked']).head()
```

```python
# Label Encoding (Scikit-Learn)
le = LabelEncoder()
data['sex_encoded'] = le.fit_transform(data['sex'])
data[['sex', 'sex_encoded']].head()
```

```python
# 수동 매핑 (순서가 있는 경우)
class_map = {'First': 3, 'Second': 2, 'Third': 1}
data['class_score'] = data['class'].map(class_map)
data.head()
```
