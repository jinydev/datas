# 03. 피처 엔지니어링
구간화(Binning), 파생변수 생성, 문자열 데이터 가공.

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

# 1. 나이 구간화 (Binning)
df['age_bin'] = pd.cut(df['age'], bins=[0, 15, 30, 50, 100], labels=['Child', 'Young', 'Adult', 'Senior'])
df[['age', 'age_bin']].head()
```

```python
# 2. 문자열 호칭 추출 (RegEx)
df['title'] = df['name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df['title'].value_counts()
```
