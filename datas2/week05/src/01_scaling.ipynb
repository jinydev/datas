# 01. 수치형 데이터 변환 (Scaling)
StandardScaler와 MinMaxScaler의 비교 실습.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

df = sns.load_dataset('titanic')
data = df[['age', 'fare']].dropna()

# 원본 산점도
plt.scatter(data['age'], data['fare'])
plt.title('Original')
plt.show()
```

```python
# 표준화 (StandardScaler)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
df_scaled = pd.DataFrame(scaled_data, columns=['age', 'fare'])

plt.scatter(df_scaled['age'], df_scaled['fare'], color='red')
plt.title('StandardScaled')
plt.grid()
plt.show()
```

```python
# 데이터 통계 확인 (평균 0, 표준편차 1)
df_scaled.describe()
```
