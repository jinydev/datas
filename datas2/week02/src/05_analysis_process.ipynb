# 05. 분석 파이프라인
함수를 사용하여 데이터 수집, 정제, 시각화 과정을 모듈화하기.

```python
import random
import matplotlib.pyplot as plt

# 1단계: 수집 함수 정의
def get_sensor_data(n):
    data = []
    for _ in range(n):
        if random.random() < 0.1: # 10% 에러 확률
            data.append(-1)
        else:
            data.append(random.randint(0, 100))
    return data

# 2단계: 전처리 함수 정의
def clean_data(raw_data):
    clean = []
    for x in raw_data:
        if x != -1:
            clean.append(x)
    return clean

# 3단계: 시각화 함수 정의
def plot_data(data, title="Data"):
    plt.figure(figsize=(8, 4))
    plt.plot(data, marker='o', label='Sensor Value')
    plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
# 파이프라인 실행
print("1. 수집...")
raw = get_sensor_data(50)
print(f"Raw: {raw[:10]}")

print("\n2. 정제...")
processed = clean_data(raw)
print(f"Clean: {processed[:10]}")

print("\n3. 시각화...")
plot_data(processed, "Cleaned Sensor Stream")
```
