랜덤 워크 알고리즘을 구현하고 제너레이티브 아트 만들기.

```python
import matplotlib.pyplot as plt
import random

# 18. 랜덤 워크 함수
def generate_random_walk(steps):
    y = [0] 
    for _ in range(steps):
        step = 1 if random.random() > 0.5 else -1
        y.append(y[-1] + step)
    return y

# 11. 다중 워커 시뮬레이션
plt.figure(figsize=(10, 6))

for i in range(10):
    path = generate_random_walk(1000)
    plt.plot(path, alpha=0.6, linewidth=1)

plt.title("Random Walk Simulation (10 Walkers)")
plt.axhline(0, color='red', linestyle='--')
plt.grid(True, alpha=0.5)
plt.show()
```

```python
# 15. 2D 랜덤 워크 (보너스)
x_path = generate_random_walk(500)
y_path = generate_random_walk(500)

plt.figure(figsize=(6, 6))
plt.plot(x_path, y_path, color='purple', alpha=0.7)
plt.scatter(0, 0, color='red', label='Start')
plt.scatter(x_path[-1], y_path[-1], color='blue', label='End')
plt.title("2D Motion Trace")
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
