```python
# 18. 응용 예제: 데이터 크기 추산
row_count = 5_000_000 
bytes_per_row = 256 

total_bytes = row_count * bytes_per_row
total_mb = total_bytes / (1024 * 1024) 
total_gb = total_mb / 1024 

print("--- 메모리 견적 ---")
print(f"예상 크기: {total_mb:.2f} MB ({total_gb:.4f} GB)")
```
