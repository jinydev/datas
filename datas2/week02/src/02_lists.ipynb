# 02. 데이터 컨테이너 (리스트)
데이터의 집합을 다루고 시각화하는 방법.

```python
import matplotlib.pyplot as plt

# 2. 리스트 생성
grades_sem1 = [85, 90, 88, 75, 92]

# 5. 인덱싱
print(f"첫 점수: {grades_sem1[0]}")
print(f"마지막 점수: {grades_sem1[-1]}")

# 11. 데이터 추가
grades_sem1.append(95)
print(f"업데이트: {grades_sem1}")
```

```python
# 15. range() 함수
weeks = list(range(1, len(grades_sem1) + 1))

# 13. 시각화
plt.figure(figsize=(8, 4))
plt.plot(weeks, grades_sem1, marker='o', label='1학기')

# 17. 리스트 병합 및 추가 시각화
grades_sem2 = [80, 85, 90]
all_grades = grades_sem1 + grades_sem2
all_weeks = list(range(1, len(all_grades) + 1))

plt.plot(all_weeks, all_grades, linestyle='--', color='gray', label='전체 흐름')

plt.title("Academic Performance Trend")
plt.legend()
plt.grid(True)
plt.show()
```
