# 01. 비극의 데이터 (타이타닉)
Seaborn 내장 데이터셋을 로드하고 구조 파악하기.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
df = sns.load_dataset('titanic')
df.head()
```

```python
# 결측치 확인
df.info()
```

```python
# 나이 분포 시각화
plt.figure(figsize=(8, 4))
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()
```
