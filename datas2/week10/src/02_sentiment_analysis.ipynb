# 02. 감성 분석 (Sentiment Analysis)
네이버 영화 리뷰 데이터를 이용한 긍정/부정 분류 모델 (Pipeline 활용).

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

# 데모 데이터
train_data = [
    ("이 영화 진짜 재밌다", 1),
    ("시간 아깝다 비추", 0),
    ("연기가 너무 좋았어요", 1),
    ("지루하고 졸림", 0)
]
df = pd.DataFrame(train_data, columns=['text', 'label'])

# 파이프라인 구축
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', LogisticRegression())
])

pipeline.fit(df['text'], df['label'])
```

```python
# 예측
reviews = ["완전 재밌어요", "노잼입니다"]
pipeline.predict(reviews)
```

**Note**: 실제 한국어 처리를 위해서는 `KoNLPy` 설치와 전처리가 필요합니다.
