# 01. 텍스트 마이닝 기초
BoW, TF-IDF, 코사인 유사도, 워드클라우드 실습.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 예제 데이터
docs = [
    'python is fun',
    'machine learning is fun',
    'python is for data science'
]

# 1. BoW (Count)
vect = CountVectorizer()
count_matrix = vect.fit_transform(docs)
print("Vocabulary:", vect.get_feature_names_out())
print("Matrix:\n", count_matrix.toarray())
```

```python
# 2. TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(docs)
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
```

```python
# 3. WordCloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(background_color='white').generate(" ".join(docs))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
