# 5.2.2 초기 데이터 탐색 (EDA) 실습

## [실습 1] 1단계: `df.head()` - 서류 봉투 살짝 열어보기

```python
import seaborn as sns

# 타이타닉 데이터를 표(DataFrame) 형태로 불러옵니다.
df = sns.load_dataset('titanic')

print("데이터 상위 5줄 미리보기:")
print(df.head())
```

## [실습 2] 2단계: `df.info()` - 엑스레이 (결측치 검사)

```python
print("\n--- 데이터 결측치 및 타입 확인 ---")
df.info()
```

## [실습 3] 3단계: `df.describe()` - 숫자의 통계적 요약

```python
print("\n--- 수치형 데이터 통계 요약 ---")
print(df.describe())
```
