# 6.4.1 데이터프레임 종합 건강검진 실습

## [실습 0] 준비물: 외부 서울시 교통사고 데이터 불러오기

```python
import pandas as pd

# CSV 파일 읽기 (인코딩을 한국어 설정인 euc-kr로 맞춤)
# 데이터 경로는 쥬피터 노트북 실행 위치에 따라 달라질 수 있습니다.
df = pd.read_csv('data/2016-01-2016-12_Seoul_Accident.csv', encoding='euc-kr', index_col=0)

# 처음 5줄만 가볍게 살펴보기 (head)
print("--- 🔬 처음 5줄 미리보기 ---")
print(df.head())
```

## [실습 1] 가장 중요한 엑스레이 촬영: .info()

```python
print("--- 🩺 데이터 구조 엑스레이 촬영 ---")
df.info()
```

## [실습 2] 데이터의 대략적인 크기와 수치 요약: .describe()

```python
print("--- 📊 기초 통계량 요약 리포트 ---")
summary = df.describe()
print(summary)
```

## [실습 3] 꿀팁: 행과 열 거꾸로 돌려보기 (Transpose)

```python
print("--- 🔄 피벗 스윙! 행과 열 뒤집기 ---")
# describe() 의 결과를 뒤집어서 보기 쉽게 바꿉니다.
print(df.describe().T)
```
