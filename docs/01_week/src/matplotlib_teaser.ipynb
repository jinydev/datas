# 1주차 3강: 데이터 시각화 맛보기 (Teaser Examples)

이 노트북은 파이썬 기초 문법과 Matplotlib을 활용하여 다양한 그래프를 그려보는 실습 예제입니다.
복잡한 문법보다는 **"데이터가 그림으로 바뀌는 경험"**에 집중해 주세요.

## 1. 간단한 선 그래프 (Line Plot)
가장 기본적인 선 그래프를 그려봅니다.

```python
import matplotlib.pyplot as plt

# 1. 데이터 준비
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 40]

# 2. 그래프 그리기
plt.plot(x, y)
plt.title("My First Plot")
plt.show()
```

## 2. 막대그래프 (Bar Chart)
요일별 판매량을 막대그래프로 표현해 봅니다.

```python
# 데이터 정의
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales = [50, 60, 55, 70, 90]

# 막대그래프 그리기 (color='brown'으로 커피색 지정)
plt.bar(days, sales, color='brown')

# 제목 및 축 레이블 설정
plt.title("Coffee Sales by Day")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

## 3. 산점도 (Scatter Plot)
키와 몸무게 데이터를 랜덤으로 생성하여 상관관계를 확인합니다.

```python
import random

# 랜덤 데이터 50개 생성
heights = [random.randint(150, 190) for _ in range(50)]
weights = [random.randint(50, 100) for _ in range(50)]

# 산점도 그리기
plt.scatter(heights, weights)

plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

## 4. 원 그래프 (Pie Chart)
비율을 한눈에 볼 수 있는 원 그래프를 그립니다.

```python
# 데이터 정의
fruits = ['Apple', 'Banana', 'Strawberry', 'Grape']
votes = [30, 20, 35, 15]

# 원 그래프 그리기 (autopct로 비율 표시)
plt.pie(votes, labels=fruits, autopct='%1.1f%%')

plt.title("Favorite Fruits")
plt.show()
```
