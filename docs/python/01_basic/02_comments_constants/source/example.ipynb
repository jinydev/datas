```python
def calculate_area(radius):
    """
    원의 넓이를 계산하는 함수입니다.

    Args:
        radius (float): 원의 반지름 값
        
    Returns:
        float: 계산된 원의 넓이 단위
    """
    result = 3.14 * radius * radius
    return result
```
